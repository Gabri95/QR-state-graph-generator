```python
from qr_model import *

file_name = 'complete_graph.png'

variables = [
    #Variable('I', ['0', '+', 'max'], [False, True, False]),
    Variable('I', ['0', '+'], [False, True]),
    Variable('V', ['0', '+', 'max'], [False, True, False]),
    Variable('H', ['0', '+', 'max'], [False, True, False]),
    Variable('P', ['0', '+', 'max'], [False, True, False]),
    Variable('O', ['0', '+', 'max'], [False, True, False])
]

model = Model(variables)

model.addVConstraint(lambda variables : (variables['O'].val == 2 and variables['V'].val == 2) or (variables['O'].val != 2 and variables['V'].val != 2))
model.addVConstraint(lambda variables : (variables['O'].val == 0 and variables['V'].val == 0) or (variables['O'].val != 0 and variables['V'].val != 0))

model.addVConstraint(lambda variables : (variables['O'].val == 2 and variables['P'].val == 2) or (variables['O'].val != 2 and variables['P'].val != 2))
model.addVConstraint(lambda variables : (variables['O'].val == 0 and variables['P'].val == 0) or (variables['O'].val != 0 and variables['P'].val != 0))

model.addVConstraint(lambda variables : (variables['O'].val == 2 and variables['H'].val == 2) or (variables['O'].val != 2 and variables['H'].val != 2))
model.addVConstraint(lambda variables : (variables['O'].val == 0 and variables['H'].val == 0) or (variables['O'].val != 0 and variables['H'].val != 0))

model.addVConstraint(lambda variables : (variables['V'].val == 2 and variables['P'].val == 2) or (variables['V'].val != 2 and variables['P'].val != 2))
model.addVConstraint(lambda variables : (variables['V'].val == 0 and variables['P'].val == 0) or (variables['V'].val != 0 and variables['P'].val != 0))

model.addVConstraint(lambda variables : (variables['V'].val == 2 and variables['H'].val == 2) or (variables['V'].val != 2 and variables['H'].val != 2))
model.addVConstraint(lambda variables : (variables['V'].val == 0 and variables['H'].val == 0) or (variables['V'].val != 0 and variables['H'].val != 0))

model.addVConstraint(lambda variables : (variables['P'].val == 2 and variables['H'].val == 2) or (variables['P'].val != 2 and variables['H'].val != 2))
model.addVConstraint(lambda variables : (variables['P'].val == 0 and variables['H'].val == 0) or (variables['P'].val != 0 and variables['H'].val != 0))

model.addPConstraint(True, 'V', 'H')
model.addPConstraint(True, 'H', 'P')
model.addPConstraint(True, 'P', 'O')

model.addIConstraint(True, 'I', 'V')
model.addIConstraint(False, 'O', 'V')
```

```python
for v in model.variables:
    print(v.name)
    for var, incr in v.p_source.items():
        print('\tP' + ('+' if incr else '-') + ' -> ' + var )
    
    for var, incr in v.i_source.items():
        print('\tI' + ('+' if incr else '-') + ' -> ' + var )

for v in model.variables:
    print(v.name)
    for var, incr in v.p_target.items():
        print('\tP' + ('+' if incr else '-') + ' <- ' + var )
    
    for var, incr in v.i_target.items():
        print('\tI' + ('+' if incr else '-') + ' <- ' + var )
        
```

```python

state_graph = model.envisioning(np.array([0, 0, 0, 0, 0], dtype=int), np.array([0, 0, 0, 0, 0], dtype=int), 'I')
#state_graph = model.envisioning(np.array([0, 0, 0, 0, 0], dtype=int), np.array([0, 0, 0, 0, 0], dtype=int), 'I')
```

```python
print(len(state_graph.nodes()))
print(len(state_graph.edges()))

state_graph.layout(prog='dot')


state_graph.draw(file_name)
```
